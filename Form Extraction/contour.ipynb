```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("baseimage.jpg")
img = cv2.resize(img,(800,1000))
cv2_imshow(img)
print(img.shape)
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
imageGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(imageGray)
```

```python
imageCanny = cv2.Canny(imageGray,100,100) #edge detector
cv2_imshow(imageCanny)
```

```python
imageDilation = cv2.dilate(imageCanny,kernel,iterations=1)
cv2_imshow(imageDilation)
```
