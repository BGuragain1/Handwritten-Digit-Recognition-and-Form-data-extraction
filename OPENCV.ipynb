```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#read images
image = cv2.imread("image.jpg")
kernel = np.ones((5,5),np.uint8)
```

```python
cv2_imshow(image)

#converting to grey
imageGray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(imageGray)

#blurring
imageBlur = cv2.GaussianBlur(imageGray,(7,7),0)
cv2_imshow(imageBlur)

#detecting edges
imageCanny = cv2.Canny(imageGray,100,100) #edge detector
cv2_imshow(imageCanny)

#increase thickness of the edges, iteration means how much time kernel would move around the image
imageDilation = cv2.dilate(imageCanny,kernel,iterations=1)
cv2_imshow(imageDilation)

#decrease the thickness of edge
imageEroded = cv2.erode(imageDilation,kernel,iterations=1)
cv2_imshow(imageEroded)
```

```python
#resizing and cropping
print(image.shape)

#resizing the image
imageresize = cv2.resize(image,(500,400))
cv2_imshow(imageresize)

#cropping the image
imageCropped = image[100:200,300:500] #height first and width here
print("Cropped image:")
cv2_imshow(imageCropped)
```

```python
#shapes and texts
#draw on images and put texts in images
image = np.zeros((512,512,3),np.uint8)

# #coloring the image of specific part
# image[200:300,100:300] = 255,0,0

# #coloring whole image
# image[:] = 255,0,0

#create line
#starting point, ending point of line , color , thickess inside the function parameters
cv2.line(image,(0,0),(image.shape[1],image.shape[0]),(0,255,0),3)

#create rectange
cv2.rectangle(image,(0,0),(250,350),(0,0,255),cv2.FILLED)

#cerating circles
cv2.circle(image,(400,50),30,(255,255,0),5)

#putting text
cv2.putText(image," OPEN CV ",(300,200),cv2.FONT_HERSHEY_COMPLEX,1,(0,150,0),2)
cv2_imshow(image)
```

```python
#WARP Perspective
img = cv2.imread("cards.jpg")

width,height = 250,350
#declaring points to get specific points of image
pts1 = np.float32([[111,219],[287,188],[154,482],[352,440]])
#declaring which points are from which part like which positions
pts2 = np.float32([[0,0],[width,0],[0,height],[width,height]])

matrix = cv2.getPerspectiveTransform(pts1,pts2)
outputImage = cv2.warpPerspective(img,matrix,(width,height))

cv2_imshow(outputImage)
cv2_imshow(img)
```

```python
#joining images
#useful when there are lot of images and we are running it again and again
hor = np.hstack((image,image))
ver = np.vstack((image,image))

cv2_imshow(hor)
cv2_imshow(ver)
```
